```python
df_by_season = bikes.groupby("season")
```

```python
type(df_by_season)
```

```python
s_casual = df_by_season.casual.sum()
```

```python
s_registered = df_by_season.registered.sum()
```

```python
df_by_season
```

```python
frames = [s_casual, s_registered]
```

```python
df = pd.concat(frames, axis =1)
```

```python
df
```

```python
df["total"]= df.casual + df.registered
```
